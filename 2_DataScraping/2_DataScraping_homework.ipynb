```python
driver = webdriver.Chrome()
driver.get(url)

#Creat list
project_list = []
category_location_list=[]
link_list=[]
pdf_list=[]  

sleep(3)

#Choose search elements
menu_button = driver.find_element("xpath","//*[@id='js-afd-mobile__btn-menu']").click()

sleep(2)

professional_button = driver.find_element("xpath","//*[@id='hamburger-menu-mobile']/div[1]/div[2]/ul/li[5]").click()

sleep(3)

search_query = driver.find_element("id","searchbar-input")

search_query.send_keys("Foster + Partners")
search_query.send_keys(Keys.RETURN)

sleep(3)

foster_button=driver.find_element("xpath","/html/body/div[4]/div/div[3]/div/div[1]/div/div[2]/div[1]/div[2]/h2/a").click()

sleep(3)

#Creat a function in order to select the project category

def extract_page_data(driver):

    projects=driver.find_elements("class name", "office__categories-category")

    print(len(projects))

    sleep(5)

    for p in range(len(projects)):

        #Find title per project
        title = projects[p].find_element("class name","office__categories-category-title").text        
        print(title)
        
        #Find category and location
        category_location = projects[p].find_element("class name","office__categories-category-city").text
        print(category_location)
        
        #Find the link
        link = projects[p].find_element("tag name","a").get_attribute('href')
        print(link)
        
        #Add items to the existing lists
        project_list.append(title)
        category_location_list.append(category_location)
        link_list.append(link)
      
   
```

```python
driver = webdriver.Chrome()
driver.get(url)

#Creat list
project_list = []
category_location_list=[]
link_list=[]
pdf_list=[]  

sleep(3)

#Choose search elements
menu_button = driver.find_element("xpath","//*[@id='js-afd-mobile__btn-menu']").click()

sleep(2)

professional_button = driver.find_element("xpath","//*[@id='hamburger-menu-mobile']/div[1]/div[2]/ul/li[5]").click()

sleep(3)

search_query = driver.find_element("id","searchbar-input")

search_query.send_keys("Foster + Partners")
search_query.send_keys(Keys.RETURN)

sleep(3)

foster_button=driver.find_element("xpath","/html/body/div[4]/div/div[3]/div/div[1]/div/div[2]/div[1]/div[2]/h2/a").click()

sleep(3)


# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down 
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    sleep(3)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    if new_height == last_height:
        break
    last_height = new_height



#Creat a function in order to select the project category

extract_page_data(driver)

driver.quit()
```

```python
import csv

with open('archname.csv', 'w',newline='',encoding='utf-8-sig') as csvfile: 
    writer = csv.writer(csvfile) 
    for name in project_list:
        print(name)
        writer.writerow([name])
```

```python
with open('archcategory.csv', 'w',newline='',encoding='utf-8-sig') as csvfile: 
    writer = csv.writer(csvfile) 
    for category in category_location_list:
        print(category)
        writer.writerow([category])
```

```python
with open('archlink.csv', 'w',newline='') as csvfile: 
    writer = csv.writer(csvfile) 
    for link in link_list:
        print(link)
        writer.writerow([link])
```

```python
data=zip(project_list, category_location_list, link_list)

with open('archdailydata.csv', 'w', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)
    project_number=0
    
    for project, category, link in data:
        project_number+=1
        writer.writerow([f'Project {project_number}:',project])
        writer.writerow(['Category:',category])
        writer.writerow(['Link:',link])
        writer.writerow([])
        
```

```python
## Iterate through the csv and download the images
#import library
import requests
import os

driver = webdriver.Chrome()
output_folder = "images"
i=0

#read the produced csv and iterate through the list
with open("archlink.csv", 'r',encoding='utf-8') as file:
    csvreader = csv.reader(file)    
    for url in csvreader:       
                
        #open each link 
        driver.get(url[0])
        sleep(3)
        
        #find the image link
        imglink = driver.find_element("tag name","source").get_attribute("srcset")        
        
        print(imglink)
        i += 1            
        response = requests.get(imglink)
        #save the images
        
        imagename = os.path.join(output_folder, "Project"+"_"+ str(i) + ".jpg")
        with open(imagename, "wb") as f:
            f.write(response.content)
       
    
#Close the webpage
driver.quit()
```
